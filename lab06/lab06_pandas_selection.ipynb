## Ćwiczenie 2: Filtrowanie — pytania menedżera restauracji

```python
# "Ile rachunków mieliśmy w piątek?"
ile_piatek= (tips['day']=='Fri').sum() 
print(f'W piątek było {ile_piatek} rachunków.')
```

```python
# "Ile rachunków złożyły grupy 1-osobowe? A 6-osobowe?"
jedno_os=(tips['size']==1).sum()
szescio_os=(tips['size']==6).sum()
print(f'Grupy 1-osobowe złożyły {jedno_os} rachunków, natomiast 6-osobowe {szescio_os}.')
```

```python
## "Jaki jest najwyższy rachunek od kobiety?"
najwyzszy_k=tips.loc[tips['sex']=='Female','total_bill'].max()
print(f'Najwyższy rachunek od kobety wynosi {najwyzszy_k}.')
```

```python
# "Ilu niepalących mężczyzn jadło u nas obiad w weekend?"
# (sex == Male) & (smoker == No) & (time == Dinner) & (day isin Sat/Sun)
niepalacy_m=tips[(tips['sex']=='Male') & (tips['smoker']=='No') & (tips['time']=='Dinner') & (tips['day'].isin (['Sat', 'Sun']))].shape[0]
print(f'W weekend jadło {niepalacy_m} niepalących mężczyzn.')
```

```python
# "Podaj średni rachunek dla dużych grup (4+) w czwartek."
duza_gr=tips.loc[(tips['size']>=4) & (tips['day']=='Thur'),'total_bill'].mean()
print(f'Średni rachunek dla dużych grup wynosi {duza_gr}')
```

```python
# "Ilu klientów dało napiwek powyżej 25% przy rachunku poniżej 10$?"
# To są podejrzanie hojni — sprawdź czy to nie błędy w danych.
pow_25=((tips['tip_pct']>25) & (tips['total_bill']< 10)).sum()
print(f'{pow_25} klientów dało napiwek powyżej 25% przy rachunku poniżej 10$')
```

Wysokie napiwki przy niskich rachunkach to prawdopodnie efekt zaokrąglenia kwot przez klientów.

```python
# "Rachunki z weekendu (Sat, Sun) w zakresie 15-30$ — ile ich jest?"
weekend_filtr=((tips['day'].isin(['Sat', 'Sun'])) & (tips['total_bill'].between(15,30))).sum()
print(f'Rachunków z weekednu z zakresu 15-30$ jest {weekend_filtr}.')

weekend_query=tips.query("day in ['Sat', 'Sun'] & 15<= total_bill<= 30")
l_rachunkow=len(weekend_query)
print(f'Rachunków z weekednu z zakresu 15-30$ jest {l_rachunkow}')
```

```python
#"Ile rachunków od niepalących mężczyzn na obiedzie przekracza średnią?"
# Użyj query z @srednia
srednia = tips['total_bill'].mean()
ile_rachunkow=tips.query("sex=='Male' & smoker=='No' & time=='Dinner' & total_bill > @srednia")
l_rachunkow_sr=len(ile_rachunkow)
print(f'{l_rachunkow} rachuni od niepalących mężczyzn na obiedzie przekraczają średnią')
```
