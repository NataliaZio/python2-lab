# 1a

```python
dane_pensja={'Anna': 5500,
'Jan': 7200,
'Ewa': 4800,
'Marek': 9100,
'Kasia': 6300
}
pensja = pd.Series(dane_pensja)
print(pensja)
```

# 1b

```python
najwiecej_zarabia = pensja.idxmax()
print(f'Najwięcej zarabia: {najwiecej_zarabia}.')
srednia = pensja.mean()
print(f'Średnia pensja wynosi: {srednia}.')

pow_sr = pensja[pensja>srednia]
print(f'Powyej średniej zarabiają: \n{pow_sr}')

ponizej_sr = pensja[pensja<6000]
klucze = len(ponizej_sr.keys())
print(f'{klucze} osób zarabia ponizej 6000zł')

```

# 1c

```python
po_podwyzce = (pensja * 1,10)
print(f'Pensje po podwyce o 10% wynoszą: \n{po_podwyzce}')

premia = pensja + 500 
print(f'Pensje po premii 500zł: \n{premia}')

nowa_pensja = (pensja * 1.10) + 500
print(f'Nowa pensja wynosi: \n{nowa_pensja}')

posortowane = nowa_pensja.sort_values(ascending=False)
print(f'Ranking nowych pensji od najwyszej: \n{posortowane}')

```

Dla kolumn tekstowych dodaje statystyki, których nie da się wyliczyc matematycznie: uniqe, top, freq. Pozwala to szybko wyróznić dominujące grupy.

# 3c

```python
podzial= penguins['species'].value_counts()
print(f'Ile pingwinów kadego gatunku?: \n{podzial}')

wyspa= penguins['island'].value_counts()
print(f'\n Ile pingwinów na kadej wyspie?: \n{wyspa}')

rozklad_plci= penguins['sex'].value_counts() 
print(f'\nRozkład płci: \n{rozklad_plci}')
```

Rozkad płci jest zbalansowany: samce - 168, samice - 165.

```python
wyniki= penguins['species'].value_counts(normalize=True)*100
print(f"Gatunek Adelie stanowi {wyniki['Adelie']} %")
```

# Ćw.4

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

# 4a

```python
sr_rachunek= tips['total_bill'].mean()
print(f'Średni rachunek wynosi: {sr_rachunek:.2f}')
sr_napiwek_pr= tips['tip_pct'].mean()
print(f'\n Średni napiwek wynosi: {sr_napiwek_pr:.2f}%')
dzien= tips['day'].value_counts().idxmax()
print(f'\nW {dzien} występuje najwięcej rachunków')
palacze=tips['smoker'].value_counts()['Yes']
print(f'\n{palacze} rachunków jest od palaczy.')
mediana_napiwkow= tips['tip'].median()
print(f'\nMediana napiwku wynosi: {mediana_napiwkow}.')
podzial_czasu= tips['time'].value_counts(normalize=True)*100
print(f'\nProcent rachunków z obiadu: {podzial_czasu['Dinner']:.2f}%.')
rozmiary=tips['size'].nunique()
print(f'\nW danych występuje {rozmiary} unikalnych rozmiarów grup.')


```
