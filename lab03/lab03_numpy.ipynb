# Produkty w sklepie internetowym

```python
nazwy = np.array(['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka'])
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
print(f"Średnia cena: {ceny.mean()} zł")
```

```python
print(f"Najdroższy produkt (cena): {ceny.max()} zł")
```

```python
print(f"Najtańszy produkt (cena): {ceny.min()} zł")
```

```python
wartosc_produktow = ceny * ilosci_magazyn
laczna_wartosc = wartosc_produktow.sum()
print(f"Łączna wartość magazynu: {laczna_wartosc} zł")
```

```python
ceny_rabat = ceny * 0.85  
ceny_vat = ceny * 1.23    
powyzej_1000 = (ceny > 1000).sum() 
print(f"Liczba produktów powyżej 1000 zł: {powyzej_1000}")
```

```python
maska_oceny = oceny_klientow >= 4.5
print("Produkty z wysoką oceną:")
print(f"Nazwy: {nazwy[maska_oceny]}") 
print(f"Ceny: {ceny[maska_oceny]}")

maska_tanie = ceny < 200 
print("\nProdukty poniżej 200 zł:")
print(f"Nazwy: {nazwy[maska_tanie]}")
```

```python
maska_magazyn = ilosci_magazyn < 30 
print(f"Niski stan: {nazwy[maska_magazyn]}") 

wartosci = ceny * ilosci_magazyn 
wartosc_niskich = wartosci[maska_magazyn].sum() 
print(f"Łączna wartość tych produktów: {wartosc_niskich} zł")

suma_niskie = ilosci_magazyn[maska_magazyn].sum()
suma_wszystkie = ilosci_magazyn.sum()
procent = (suma_niskie / suma_wszystkie) * 100

print(f"Ilościowo produkty te stanowią {procent:.2f}% całego magazynu.")
```

```python
maska_okazja = (oceny_klientow >= 4.5) & (ceny < 1000)
print(f"Produkty będące okazją: {nazwy[maska_okazja]}")
print(f"Ich ceny: {ceny[maska_okazja]} zł, oceny: {oceny_klientow[maska_okazja]}")


maska_dobra_ocena = oceny_klientow >= 4.0
maska_slaba_ocena = oceny_klientow < 4.0

srednia_dobra = ceny[maska_dobra_ocena].mean()
srednia_slaba = ceny[maska_slaba_ocena].mean()

print(f"\nŚrednia cena produktów z oceną >= 4.0: {srednia_dobra:.2f} zł")
print(f"Średnia cena produktów z oceną < 4.0: {srednia_slaba:.2f} zł")
```

```python
import time
```

```python
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
