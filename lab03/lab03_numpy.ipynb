```python
import time
```

```python
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])

produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

# Zadanie 3a: Analiza wynagrodzeń

```python
srednia = pensje.mean()
mediana = np.median(pensje)
std = pensje.std()

print(f'Średnia: {srednia} zł, Mediana: {mediana} zł, Odchylenie: {std:.2f} zł')

nazwy_np = np.array(pracownicy)
powyzej_sredniej = nazwy_np[pensje > srednia]
print(f'Powyżej średniej zarabiają: {powyzej_sredniej}')

ponizej_mediany = nazwy_np[pensje < mediana]
print(f'Poniżej mediany zarabiają: {ponizej_mediany}')

rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)
print(f'Rozstęp: {rozstep} zł, IQR: {iqr} zł')
```

# Zadanie 3b:

```python
warunki_premii = [
    oceny >= 4.5,
    (oceny >= 4.0) & (oceny < 4.5),
    oceny < 4.0
]
wartosci_premii = [2000, 1000, 0]

premie = np.select(warunki_premii, wartosci_premii, default=0) 
nowe_pensje = pensje + premie
print(f'Przyznane premie: {premie}')
print(f'Nowe wynagrodzenia całkowite: {nowe_wynagrodzenie}')

max_p = premie.max()
print(f'Osoby z najwyższą premią ({max_p} zł): {nazwy_np[premie == max_p]}')
```

# Zadanie 3c:

```python
kolejnosc = np.argsort(pensje)[::-1]
print('Ranking pracowników (od najwyższych zarobków):')
for i, idx in enumerate(kolejnosc, 1):
    print(f'{i}. {pracownicy[idx]:<6} | Pensja: {pensje[idx]} zł | Staż: {staz[idx]} lat')
```

# Zadanie 3d:

```python
kor_staz_pensja = np.corrcoef(staz, pensje)[0, 1]
kor_staz_ocena = np.corrcoef(staz, oceny)[0, 1]
print(f'Korelacja staż-pensja: {kor_staz_pensja:.2f}')
print(f'Korelacja staż-ocena: {kor_staz_ocena:.2f}')

if abs(kor_staz_pensja) > abs(kor_staz_ocena):
    print('Silniejszy związek występuje między stażem a pensją.')
else:
    print('Silniejszy związek występuje między stażem a oceną.')
```

```python
print(f'Wiek - średnia: {wiek.mean():.1f}, mediana: {np.median(wiek):.1f}') 
print(f'Wydatki - max: {wydatki.max():.2f}, min: {wydatki.min():.2f}') 

kor_wiek_wydatki = np.corrcoef(wiek, wydatki)[0, 1] 
print(f'Korelacja wiek-wydatki: {kor_wiek_wydatki:.3f} (powinna być bliska 0)')

sr = wydatki.mean() 
sd = wydatki.std()

segmenty = np.where(wydatki > sr + sd, 'WYSOKI', 
           np.where(wydatki < sr - sd, 'NISKI', 'ŚREDNI'))

nazwy_seg, ilosc = np.unique(segmenty, return_counts=True)
for n, i in zip(nazwy_seg, ilosc):
    print(f'Segment {n}: {i} klientów')
```
